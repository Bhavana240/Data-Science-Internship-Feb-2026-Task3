```python
# Problem 1: Employee Performance Bonus Eligibility

# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Identify employees with the highest score (handle ties)
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation (if any)
import string
for char in string.punctuation:
    query = query.replace(char, "")

    # Step 3: Split into words
    words = query.split()

    # Step 4: Count frequency of each word
    word_count = {}

    for word in words:
        word_count[word] = word_count.get(word, 0) + 1

        # Step 5: Filter words searched more than once
        frequent_words = {}

        for word, count in word_count.items():
            if count > 1:
                    frequent_words[word] = count

                    # Step 6: Display result
                    print(frequent_words)
```

```python
# Problem 3: Sensor Data Validation

# List of sensor readings (index = hour)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid (even) readings
valid_readings = []

# Step 1: Iterate through readings with index
for index, value in enumerate(sensor_readings):
    # Step 2: Check if reading is even
        if value % 2 == 0:
                valid_readings.append((index, value))

                # Step 3: Display valid sensor readings
                print("Valid Sensor Readings (Hour, Value):", valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

# List of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Step 1: Extract domain from each email
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate total emails
total_emails = len(emails)

# Step 3: Calculate and display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

# List of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Define spike threshold (30% above average)
threshold = average_sales * 1.30

# Step 3: Detect sales spikes
for index, value in enumerate(sales):
    if value > threshold:
            print(f"Day {index + 1}: {value}")
```

```python
# Problem 6: Duplicate User ID Detection

# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to count occurrences
user_count = {}

# Step 1: Count user IDs
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

    # Step 2: Display duplicates only
    for user, count in user_count.items():
        if count > 1:
                print(f"{user} → {count} times")
```
